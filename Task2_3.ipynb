Nb of sec with 6250B < Vol DL < 31250B,Number of seconds with IP Volume DL between …
Nb of sec with 6250B < Vol UL < 37500B,Number of seconds with IP Volume UL between …

Nb of sec with Vol UL < 1250B,Number of seconds with IP Volume UL <

````python
# Aggregate per user
user_aggregate = telecom_data.groupby('MSISDN/Number').agg({
    'Number of xDR sessions': 'count',
    'Session duration': 'sum',
    'Total DL': 'sum',  # Total download data
    'Total UL': 'sum',  # Total upload data
    'Social Media DL': 'sum',  # Total Social Media download data
    'Social Media UL': 'sum',  # Total Social Media upload data
    'YouTube DL': 'sum',  # Total YouTube download data
    'YouTube UL': 'sum',  # Total YouTube upload data
    'Netflix DL': 'sum',  # Total Netflix download data
    'Netflix UL': 'sum',  # Total Netflix upload data
    'Google DL': 'sum',  # Total Google download data
    'Google UL': 'sum',  # Total Google upload data
    'Email DL': 'sum',  # Total Email download data
    'Email UL': 'sum',  # Total Email upload data
    'Gaming DL': 'sum',  # Total Gaming download data
    'Gaming UL': 'sum',  # Total Gaming upload data
    'Other DL': 'sum',  # Total Other download data
    'Other UL': 'sum'   # Total Other upload data
})

print(user_aggregate.head())
````

````python
#performing principal component analysis(PCA) to reduce the size of the data

from sklearn.decomposition import PCA
import numpy as np

# Preparing data for PCA
X = user_aggregate.drop('Number of xDR sessions', axis=1)

# Standardizing the data
X_std = (X - X.mean()) / X.std()

# Performing PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_std)

# Creating DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

print(pca_df.head())
````

## **Task 3**

````python
# Convert columns to numeric
columns_to_convert = [
    'Number of xDR sessions',
    'Session duration',
    'Total DL',
    'Total UL'
]

for col in columns_to_convert:
    telecom_data[col] = pd.to_numeric(telecom_data[col], errors='coerce')

# Aggregating metrics per MSISDN
engagement_metrics = telecom_data.groupby('MSISDN/Number').agg({
    'Number of xDR sessions': 'sum',
    'Session duration': 'sum',
    'Total DL': 'sum',
    'Total UL': 'sum'
}).reset_index()

# Top 10 customers per engagement metric
top_10_sessions = engagement_metrics.sort_values(by='Number of xDR sessions', ascending=False).head(10)
top_10_duration = engagement_metrics.sort_values(by='Session duration', ascending=False).head(10)
top_10_total_dl = engagement_metrics.sort_values(by='Total DL', ascending=False).head(10)
top_10_total_ul = engagement_metrics.sort_values(by='Total UL', ascending=False).head(10)

print("Top 10 customers by number of xDR sessions:\n", top_10_sessions)
print("\nTop 10 customers by Session duration:\n", top_10_duration)
print("\nTop 10 customers by Total DL:\n", top_10_total_dl)
print("\nTop 10 customers by Total UL:\n", top_10_total_ul)
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Normalizing the engagement metrics
scaler = StandardScaler()
engagement_metrics_normalized = scaler.fit_transform(engagement_metrics.drop('MSISDN/Number', axis=1))

# Run k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['Cluster'] = kmeans.fit_predict(engagement_metrics_normalized)

# Compute metrics for each cluster
cluster_summary = engagement_metrics.groupby('Cluster').agg({
    'Number of xDR sessions': ['min', 'max', 'mean', 'sum'],
    'Session duration': ['min', 'max', 'mean', 'sum'],
    'Total DL': ['min', 'max', 'mean', 'sum'],
    'Total UL': ['min', 'max', 'mean', 'sum']
}).reset_index()

print(cluster_summary)
````

````python
# Aggregating total traffic per application
traffic_apps = ['Social Media', 'YouTube', 'Netflix', 'Google', 'Email', 'Gaming', 'Other']
for app in traffic_apps:
    engagement_metrics[f'Total {app}'] = telecom_data.groupby('MSISDN/Number')[f'{app} DL'].sum() + telecom_data.groupby('MSISDN/Number')[f'{app} UL'].sum()

# Top 10 most engaged users per application
top_10_apps = pd.DataFrame()
for app in traffic_apps:
    top_10_apps[app] = engagement_metrics.sort_values(by=f'Total {app}', ascending=False)['MSISDN/Number'].head(10).values

print(top_10_apps)
````

````python
#Finding the optimized value of k for k-means using elbow method
# Elbow method to find optimal k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(engagement_metrics_normalized)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
````
